<h1>Carbon Footprint Database</h1>
<p style="font-size: 15px">	A large portion of the book <em>How Bad Are Bananas: The Carbon Footprint of Everything</em> by Mike Berners-Lee, contained information about the carbon footprints of various items and activities. In addition to the carbon dioxide equivalent values (CO2e) provided for all listed items, in some cases, infographics like bar and line graphs were also added to further depict the carbon path of the given item. As a result, it would be useful to compile together all the data and process it in a Python script, in order to be able to compare and observe the carbon footprints of different items and processes. Taking the it a step further, it would be even more useful to have a large database containing the carbon footprints of a vast number of different things, where different carbon footprints can be compared next to each other. In addition, the program should be user-friendly or interactive, meaning a person who does not know how to code should be able to access and compare the data. As a result, the program can be made interactive through the use of a prompt, or even better, a graphic user interface (GUI).</p>

<h2>Distribution of Carbon Footprint in Various Items</h2>

```python
# Gets data of items in text files and graphs them with matplotlib.

import matplotlib.pyplot as plt


class Graphing():
    def __init__(self, file=""):
        self.file = file
        if file:
            self.raw_data = [line.rstrip() for line in open(file)]
            self.num_graphs = len(self.raw_data) // 3
        
    def draw(self, title, labels, values):
        fig1, ax1 = plt.subplots()
        ax1.pie(values, labels=labels, autopct="%1.1f%%")
        ax1.axis('equal')
        plt.title(title)
        plt.show()
        print()

    def draw_graphs(self):
        """Draw graphs based on data."""
        for i in range(self.num_graphs):
            title, labels, values = self.raw_data[i * 3: (i + 1) * 3]
            labels = labels.split(",")
            values = [float(j) for j in values.split(",")]
            self.draw(title, labels, values)
```

<h2>Interactive Prompt</h2>
<p style="font-size: 15px">Enter the number for the graph you wish to see (from 1 to 17). Enter in "q" to quit. The options available are:</p>
1. A Pint of Beer
2. Milk Bottle
3. Swimming Pool
4. A University
5. A Pair of Shoes
6. Night in a Hotel
7. New Car
8. A House
9. Loaf of Bread
10. A Quart of Gasoline
11. Carton of Eggs
12. Healthcare
13. A Person
14. The World Cup
15. A Country
16. Black Carbon
17. The World

```python
data = list(open("data.txt"))
titles = []
labels = []
values = []
graph = Graphing()

# Organize data into titles, labels and values.
for i in range(len(data)):
    if i % 3 == 0:
        titles.append(data[i].rstrip())
    elif (i + 2) % 3 == 0:
        labels.append(data[i].rstrip())
    else:
        values.append(data[i].rstrip())

# The values assigned here are for the sake of initialization.
user_input = None
g_title = 0
g_labels = 0
g_values = 0

while user_input != "q":
    user_input = input("Enter Command: ")
    
    if user_input.isdigit():
        idx = int(user_input)
        if idx in range(1, len(titles) + 1):
            g_title = titles[idx - 1]
            g_labels = labels[titles.index(g_title)].split(',')
            g_values = [float(i) for i in values[titles.index(g_title)].split(',')]
            graph.draw(g_title, g_labels, g_values)
            print()
        else:
            print("{} is not a available graph.".format(idx))
            
    elif user_input == "q":
        print("Quitting interactive prompt.")
        
    else:
        print("Command not understood.")
```
