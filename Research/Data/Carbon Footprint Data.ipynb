<h3>Distribution of Carbon Footprint in Various Items</h3>

```python
import matplotlib.pyplot as plt

raw_data = [line.rstrip() for line in open("co2e_distribution.txt")]
n = len(raw_data) // 3

for i in range(n):
    title, labels, values = raw_data[i * 3: (i + 1) * 3]
    labels = labels.split(",")
    values = [int(j) for j in values.split(",")]
    
    fig1, ax1 = plt.subplots()
    ax1.pie(values, labels=labels, autopct="%1.1f%%")
    ax1.axis('equal')
    plt.title(title)
    plt.show()
```
